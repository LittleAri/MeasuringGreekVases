#### Path-Straightening Distance

We pick the _p1_ th, and _p2_ th names from the dataset for our example of two curves. We then perform path-straightening where the number of paths is determined by the variable _splits_. This can be changed to any integer greater than 2.

```python
p1 = 0
p2 = -1

splits = 5

x1 = list(df[df['Filename']==names[p1]]['X'])
y1 = list(df[df['Filename']==names[p1]]['Y'])
x2 = list(df[df['Filename']==names[p2]]['X'])
y2 = list(df[df['Filename']==names[p2]]['Y'])
dist,a,b,c = geodDistance(x1,y1,x2,y2,k=splits)
```

#### Plot Paths

```python
fig,ax = plt.subplots(1,splits+2,figsize=(3*(splits+2),3))

ax[0].plot(x1,y1,'-b')
ax[-1].plot(x2,y2,'-b')

for i in range(0,splits):
    x = a[0,:,i]
    y = a[1,:,i]
    ax[i+1].plot(x,y,'-k')
    
for i in range(0,splits+2):
    ax[i].axis('off')
```
